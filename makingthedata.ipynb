From the above table we notice that some players have been a part of multiple teams in a season. As such their 2K ratings represent their performances as a whole in both teams and not just in a single team. To account for this, we need to create a single row for each player that takes into account the matches played for each team and modifies the statistics accordingly

```
temp3=full18
for i in full18.Name.unique():
  temp=full18[full18.Name==i]
  l=temp.index
  mpg=full18.loc[l,'MPG']
  totalshots=full18.loc[l,'FTA']+full18.loc[l,'2PA']+full18.loc[l,'3PA']
  totalnotfree=full18.loc[l,'FTA']+full18.loc[l,'2PA']
  efp=full18.loc[l,'Effective Shooting percent']
  tsp=full18.loc[l,'True shooting percent']
  gp=full18.loc[l,'GP']
  ppg=full18.loc[l,'PPG']
  rpg=full18.loc[l,'RPG']
  apg=full18.loc[l,'APG']
  spg=full18.loc[l,'SPG']
  bpg=full18.loc[l,'BPG']
  pp=0
  ap=0
  rp=0
  sp=0
  bp=0
  mp=0
  ef=0
  ts=0
  for i in l:
    pp+=ppg[i]*gp[i]
    ap+=apg[i]*gp[i]
    rp+=rpg[i]*gp[i]
    sp+=spg[i]*gp[i]
    bp+=bpg[i]*gp[i]
    mp+=mpg[i]*gp[i]
    ef+=efp[i]*totalnotfree[i]
    ts+=tsp[i]*totalshots[i]
  totalgames=sum(gp)
  totalswithfree=sum(totalshots)
  totalswithout=sum(totalnotfree)
  pp=pp/totalgames
  ap=ap/totalgames
  rp=rp/totalgames
  sp=sp/totalgames
  bp=bp/totalgames
  mp=mp/totalgames
  ef=ef/totalswithout
  ts=ts/totalswithfree
  temp3.loc[l,'GP']=totalgames
  temp3.loc[l,'PPG']=pp
  temp3.loc[l,'APG']=ap
  temp3.loc[l,'RPG']=rp
  temp3.loc[l,'SPG']=sp
  temp3.loc[l,'BPG']=bp
  temp3.loc[l,'MPG']=mp
  temp3.loc[l,'Effective Shooting percent']=ef
  temp3.loc[l,'True shooting percent']=ts
  
```
